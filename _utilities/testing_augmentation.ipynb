```python
# First import the necessary libraries
import os
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
FILL_MODE = 'nearest'

# Create the augmentation model
data_augmentation = tf.keras.Sequential([
    tf.keras.Input(shape=(150,150,3)),
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.2, fill_mode=FILL_MODE),
    tf.keras.layers.RandomTranslation(0.2, 0.2, fill_mode=FILL_MODE),
    tf.keras.layers.RandomZoom(0.2, fill_mode=FILL_MODE)
])
```

```python
def load_sample_image(img_path):
    img = tf.keras.preprocessing.image.load_img(img_path, target_size=(150, 150))
    img_array = tf.keras.preprocessing.image.img_to_array(img)
    img_array = tf.expand_dims(img_array, axis=0)  # Expand dimensions for batch
    return img_array
```

```python
# Path to your sample image
img_path = r"C:\Users\n8n0c\Documents\College\Bangkit Essentials\dataset-ml\keben-bali\keben_bali1.png"  # Replace with your image path
sample_image = load_sample_image(img_path)

# Generate augmented images
plt.figure(figsize=(10, 7))
for i in range(6):
    augmented_image = data_augmentation(sample_image, training=True)
    plt.subplot(2, 3, i + 1)
    plt.imshow(tf.cast(augmented_image[0], tf.uint8))
    plt.axis('off')

plt.tight_layout()
plt.show()
```
